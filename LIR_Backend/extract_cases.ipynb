```python
# method to extract raw case data from csv file

# standard library imports
import csv

# related third party imports
import pandas as pd
from io import StringIO

# field limit increased to avoid ParserError
csv.field_size_limit(20000000)

def extract_raw_cases():

    # path for unstructured judgment data
    input_file_path = "C:\\Users\\Charlie\\Desktop\\thesis_data\\irish_jugments_HC-CA-SC_unstructured.csv"
    output_file_path = "C:\\Users\\Charlie\\Desktop\\thesis_data\\irish_jugments_HC-CA-SC_unclean_df.csv"

    # open file
    with open(input_file_path, "r", encoding="utf8") as judgments_file:
        
        # read raw file
        raw_case_data = judgments_file.read()

        # load file data into judgments_df DataFrame
        judgments_df = pd.read_csv(StringIO(raw_case_data), sep = ";", engine = "python")

        # save judgments_df to new csv file without index
        judgments_df.to_csv(output_file_path, index = False)
        
    return judgments_df

#call extract_raw_cases:
extract_raw_cases()


 
```
